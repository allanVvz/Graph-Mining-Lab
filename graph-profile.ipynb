## Libs

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 1. Carregar a matriz de adjacência do arquivo CSV
# Supondo que os dados não possuam cabeçalho e sejam separados por vírgula.
A = np.loadtxt("data.csv", delimiter=",")

# 2. Criar o grafo G a partir da matriz.
# Se a matriz for simétrica, o grafo será não direcionado.
G = nx.from_numpy_array(A)  # cria um grafo não direcionado

# Verifica se o grafo é direcionado.
print("Grafo é direcionado?", G.is_directed())
# Exemplo de saída: Grafo é direcionado? False
```

```python
# 3. Desenhar o grafo com dimensões 12 x 7.
plt.figure(figsize=(12, 7))
pos = nx.spring_layout(G)  # cálculo de layout para melhor visualização
nx.draw(G, pos, with_labels=True, node_size=300, font_size=8)
plt.title("Grafo G")
plt.show()
```

```python
# 4. Extrair a quantidade de vértices e arestas do grafo.
num_vertices = G.number_of_nodes()
num_arestas = G.number_of_edges()
print("Quantidade de vértices:", num_vertices)
print("Quantidade de arestas:", num_arestas)
```

```python
# 5. Extrair o grau do vértice 33 e comparar com o grau médio da rede.
# Atenção: os vértices são numerados a partir de 0.
grau_v33 = G.degree(33)
graus = dict(G.degree()).values()
grau_medio = sum(graus) / num_vertices
diferenca = grau_v33 - grau_medio
print("Grau do vértice 33:", grau_v33)
print("Diferença entre o grau do vértice 33 e o grau médio da rede:", diferenca)
```

```python
# 6. Conjunto de arestas que se conectam ao vértice 33.
arestas_v33 = list(G.edges(33))
print("Arestas conectadas ao vértice 33:", arestas_v33)
```

```python
# 7. Distribuição de graus do grafo em uma figura 10 x 5.
plt.figure(figsize=(10, 5))
lista_graus = list(dict(G.degree()).values())
plt.hist(lista_graus, bins=range(min(lista_graus), max(lista_graus) + 2), align='left', rwidth=0.8)
plt.title("Distribuição de graus do grafo G")
plt.xlabel("Grau")
plt.ylabel("Frequência")
plt.show()
```
